```python
import numpy as np
import pandas as pd
import requests
import json
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
URL="https://uinames.com/api/?amount=500"
response = requests.get(URL)
results = response.json()
results
```

```python
people=pd.DataFrame(results)
people.head(20)
```

```python
people.groupby(['name','region']).sum()
```

```python
name = people.name.value_counts()
plt.figure(figsize=(10,20))
sns.barplot(x=name[:12].index,y=name[:12].values)
plt.xticks(rotation=45)
plt.title('Most common name',color = 'blue',fontsize=15)
```
