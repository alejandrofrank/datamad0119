```python
import numpy as np
import pandas as pd
import requests
import json
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

# Información sobre el API utilizado.

El siguiente API es un generador de personas. Los atributos de los mismos serán (Nombre, Apellido, Genero y región)

```python
URL="https://uinames.com/api/?amount=500"
response = requests.get(URL)
results = response.json()
results
```

```python
people=pd.DataFrame(results)
people.head()
```

# Nombre mas común

Sorprendentemente, el API arroja nombres muy variados, a excepcion de algunos nombres que se repiten. Los nombres que no se ven en la grafica no están soportados por matplotlib.

```python
name = people.name.value_counts()
plt.figure(figsize=(10,10))
sns.barplot(x=name[:12].index,y=name[:12].values)
plt.xticks(rotation=45)
plt.title('Most common name',color = 'blue',fontsize=15)
```

# Apellidos mas comunes

Al igual que en la tabla de los nombres, no se repiten mucho los apellidos.

```python
surname = people.surname.value_counts()
plt.figure(figsize=(10,10))
sns.barplot(x=surname[:12].index,y=surname[:12].values)
plt.xticks(rotation=45)
plt.title('Most common Surname',color = 'blue',fontsize=15)
```
